```python
import sys
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('/content/sample_data/30YtrYE0stU_corr.png',0)
#img = cv.imread('/content/sample_data/4D5xMPvxEGo.jpg',0)
#img = cv.imread('/content/sample_data/K2GP66Fhs6M.jpg',0)
img = cv.medianBlur(img,5)
cimg = cv.cvtColor(img,cv.COLOR_GRAY2BGR)
circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=0,maxRadius=40)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
cv2_imshow(cimg)
cv.waitKey(0)
cv.destroyAllWindows()
```
